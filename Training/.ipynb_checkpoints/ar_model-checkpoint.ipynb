 ## Autoregressive Models of order p $AR(p)$

Le modèle autorégressif est simplement une extension de la marche aléatoire. C'est essentiellement un modèle de régression qui dépend linéairement des termes précédents:$$x_t = \alpha_1x_{t-1}+...+\alpha_px_{t-p}+w_t = \sum_{i=1}^{p} t_i\alpha_ix_{t-i}+w_t$$

Il s'agit d'un modèle AR d'ordre "p", où $p$ représente le nombre de termes précédents (ou décalés) utilisés dans le modèle, $\alpha_i$ est le coefficient, et $w_t$ est un terme de bruit blanc.

L'un des aspects les plus importants du modèle d'ordre "p" est qu'il n'est pas toujours stationnaire. La stationnarité d'un modèle particulier dépend des paramètres. Par exemple, un modèle AR(1) avec $\alpha_1$ = 1 est une marche aléatoire et n'est donc pas stationnaire.

Simulons un modèle AR(1) avec $\alpha$ égal à 0,6

```python
import os
import sys
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import statsmodels.tsa.api as smt
import statsmodels.api as sm
import scipy.stats as scs
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# tsplot definition

def tsplot(y, lags=None, figsize=(15, 10), style='bmh'):
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
    with plt.style.context(style):
        fig = plt.figure(figsize=figsize)
        # mpl.rcParams['font.family'] = 'Ubuntu Mono'
        layout = (3, 2)
        ts_ax = plt.subplot2grid(layout, (0, 0), colspan=2)
        acf_ax = plt.subplot2grid(layout, (1, 0))
        pacf_ax = plt.subplot2grid(layout, (1, 1))
        qq_ax = plt.subplot2grid(layout, (2, 0))
        pp_ax = plt.subplot2grid(layout, (2, 1))

        y.plot(ax=ts_ax)
        ts_ax.set_title('Time Series Analysis Plots')
        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax, alpha=0.05)
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax, alpha=0.05)
        sm.qqplot(y, line='s', ax=qq_ax)
        qq_ax.set_title('QQ Plot')
        scs.probplot(y, sparams=(y.mean(), y.std()), plot=pp_ax)

        plt.tight_layout()
    return

# Simulate an AR(1) process with a = 0.6

np.random.seed(1)
n_samples = int(1000)
a = 0.6
x = w = np.random.normal(size=n_samples)

for t in range(n_samples):
    x[t] = a*x[t-1] + w[t]
    
_ = tsplot(x, lags=30)
```

La distribution de notre modèle AR(1) simulé est normale mais il existe une corrélation en série significative entre les valeurs décalées visibles dans les graphiques ACF et PACF.

Les tracés PACF sont utilisés pour identifier l'étendue du retard dans un mode autorégressif. Si nous ne trouvons aucune corrélation significative dans un graphique PACF après un décalage de $k$, un modèle AR(k) est généralement un bon ajustement. En examinant ce graphique, nous pouvons émettre l'hypothèse qu'un modèle AR(1) devrait s'ajuster.


#### Data Fitting

Nous pouvons maintenant ajuster un modèle AR(p) en utilisant les statsmodels de Python pour estimer le coefficient alpha et l'ordre. Si le modèle AR est correct, le coefficient alpha estimé sera proche de notre alpha réel de 0,6 et l'ordre choisi sera égal à 1.

```python
# Fit an AR(p) model to simulated AR(1) model with alpha = 0.6

mdl = sm.tsa.AR(x).fit(maxlag=30, ic='aic', trend='nc')
est_order = sm.tsa.AR(x).select_order( maxlag=30, ic='aic', trend='nc')

true_order = 1
print('\nalpha estimate: %3.5f | order_estimate %s'%(mdl.params, est_order))
print('\ntrue alpha = %s | true order = %s'%(a, true_order))
```

Simulons un processus AR(2) avec $\alpha_1$ = 0,666 et $\alpha_2$ = -0,333.

```python
# Simulate an AR(2) process

n = int(1000)
alphas = np.array([.666, -.333])
betas = np.array([0.])

# Python requires us to specify the zero-lag value which is 1
# Also note that the alphas for the AR model must be negated
# We also set the betas for the MA equal to 0 for an AR(p) model
# For more information see the examples at statsmodels.org
ar = np.r_[1, -alphas]
ma = np.r_[1, betas]

ar2 = smt.arma_generate_sample(ar=ar, ma=ma, nsample=n) 
_ = tsplot(ar2, lags=30)
```

Il existe maintenant une corrélation en série significative entre les valeurs décalées aux retards 1 et 2, comme le montre le graphique du PACF. On récupère maintenant les paramètres

```python
# Fit an AR(p) model to simulated AR(2) process

max_lag = 10
mdl = smt.AR(ar2).fit(maxlag=max_lag, ic='aic', trend='nc')
est_order = smt.AR(ar2).select_order( maxlag=max_lag, ic='aic', trend='nc')

true_order = 2
print('\ncoef estimate: %3.4f %3.4f | order estimate %s'%(mdl.params[0],mdl.params[1],est_order))
```

#### Choix du nombre de décalages

Si on se contente d'ajuster naïvement un modèle, il estimera un certain nombre de paramètres. Voir ci-dessous :

```python
mdl = smt.AR(ar2).fit()
print ('Parameters')
print (mdl.params)
print ('Standard Error')
print (mdl.bse)
```

Dans ce cas, nous savons qu'il y en a trop car nous avons simulé les données comme un processus AR(2).

Les modèles AR estimeront beaucoup plus de retards que ce n'est réellement le cas en raison de la dépendance indirecte - si $X_t$ dépend de $X_{t-1}$ qui dépend de $X_{t-2}$, alors indirectement $X_t$ dépendra de $X_{t-2}$ . En présence de plus d'un retard, les dépendances indirectes seront relevées par une simple estimation.

Cependant, nous voulons le moins de paramètres possible qui permettent d'obtenir un bon modèle, qui expliquent ce qui se passe. Tout paramètre supplémentaire et indésirable entraînera un surajustement.

L'observation du FAC et du FACP indique que seuls les deux premiers décalages peuvent être utiles. Cependant, nous allons tester empiriquement le nombre de décalages en utilisant un critère d'information quelconque (en particulier le critère d'information d'Akaike (AIC)). Python fait cela pour nous lorsque nous spécifions "maxlag" et "ic" à la fonction d'ajustement. Pour l'essayer vous-même, calculez l'AIC pour tous les modèles que nous souhaitons considérer, et notez le plus petit AIC. Ce modèle minimise la perte d'information et donne le meilleur nombre de paramètres.

```python
N = 10
AIC = np.zeros((N, 1))

for i in range(N):
    model = smt.AR(ar2)
    model = model.fit(maxlag=(i+1))
    AIC[i] = model.aic
    
AIC_min = np.min(AIC)
model_min = np.argmin(AIC)

print ('Number of parameters in minimum AIC model %s' % (model_min+1))
```

#### Évaluation des résidus

La dernière étape après l'ajustement de notre modèle consiste à évaluer son comportement résiduel (trouver un modèle adapté à notre série temporelle de telle sorte que les résidus soient des bruits blancs)

Vérifions ici la normalité des résidus.

```python
from statsmodels.stats.stattools import jarque_bera

score, pvalue, _, _ = jarque_bera(mdl.resid)

if pvalue < 0.10:
    print ('We have reason to suspect the residuals are not normally distributed.')
else:
    print ('The residuals seem normally distributed.')
```

```python
tsplot(mdl.resid, lags=30)
```

Les résidus ressemblent en effet à un bruit blanc, ce qui indique qu'on a bien expliqué les données avec notre modèle.

#### Application série financière

Modèle AR(p) sur les rendements AAPL :

```python
from backtester.dataSource.yahoo_data_source import YahooStockDataSource
startDateStr = '2006/12/31'
endDateStr = '2013/12/31'
cachedFolderName = 'yahooData/'
dataSetId = 'testPairsTrading'
instrumentIds = ['^GSPC','DOW','AAPL','MSFT', 'SPX']
ds = YahooStockDataSource(cachedFolderName=cachedFolderName,
                            dataSetId=dataSetId,
                            instrumentIds=instrumentIds,
                            startDateStr=startDateStr,
                            endDateStr=endDateStr,
                            event='history')
data = ds.getBookDataByFeature()['adjClose']
# log returns
lrets = np.log(data/data.shift(1)).fillna(0)
```

```python
lrets.AAPL.plot(figsize=(10, 8))
```

On cherche le meilleur ordre et les résidus du modèle :

```python
# Select best lag order for AAPL returns

max_lag = 30
mdl = smt.AR(lrets.AAPL).fit(maxlag=max_lag, ic='aic', trend='nc')
est_order = smt.AR(lrets.AAPL).select_order(maxlag=max_lag, ic='aic', trend='nc')

print('best estimated lag order = %s'%(est_order))

_ = tsplot(mdl.resid, lags=max_lag)
```

```python
from statsmodels.stats.stattools import jarque_bera

score, pvalue, _, _ = jarque_bera(mdl.resid)

if pvalue < 0.10:
    print ('We have reason to suspect the residuals are not normally distributed.')
else:
    print ('The residuals seem normally distributed.')
```
